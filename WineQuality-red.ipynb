## EDA With Red Wine Data
- Dataset Information: http://archive.ics.uci.edu/ml/datasets/wine+quality

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine. Due to privacy and logistic issues, only physcochemical (inputs) and sensory (the output) variables are availables (e.g. there is no data about grape types, wine brand, wine selling price etc.)

These datasets can be viewed as classification  or regression tasks. The classes are ordered and not balanced (e.g. there aremany more normal wines than excellent or poor ones). Outlier detection algorithms could be used to detect the few excellent or poor wines. Also, we are not sure if all input variables are relevent. So it could be interesting to test feature selection methods.

Attribute Information:

Input variables (based on physicochemical tests):

- 1 - fixed acidity
- 2 - volatile acidity
- 3 - citric acidity
- 4 - residual sugar
- 5 - chlorides
- 6 - free sulfur dioxide
- 7 - total sulfur dioxide
- 8 - density
- 9 - ph
- 10 - sulphates
- 11 - alcohal

Output variables (based on sensory data):

- 12 - quality (score between 0 to 10)

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
```

```python
df.quality.value_counts().plot(kind='bar')
plt.xlabel("Quality", fontsize=12)
```

```python
sns.histplot(df['fixed acidity'], kde=True)
plt.xlabel("Fixed Acidity", fontsize=12)
plt.ylabel("Count", fontsize=12)
```

```python
for i in df.columns:
    sns.histplot(df[i], kde=True)
plt.xlabel("Fixed Acidity", fontsize=12)
plt.ylabel("Count", fontsize=12)
```

```python
# Categorical plot
sns.catplot(x='quality', y='alcohol', data=df, kind='box')
plt.xlabel("Quality", fontsize=12)
plt.ylabel("Alcohol", fontsize=12)
```

```python
sns.scatterplot(x='alcohol',y='pH',hue='quality',data=df)
plt.xlabel("Alcohol", fontsize=12)
plt.ylabel("pH", fontsize=12)
```
